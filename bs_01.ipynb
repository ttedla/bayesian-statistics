# መግቢያ

የቤዚያን ስታትስቲክስ መነሻው የቤይስ ቴረም ነው። ወደ ቤዚያን ቴረም ከመሄዳችን በፊት ሁኔታዊ ዕድልን (ኮንዲሽናል ፕሮባቢሊቲን) እንመልከት። ቀጥለን ቤዚያን ቴረምን እንመለከት እና ወደ ቤዚያን ስታትስቲክስ እንሄዳለን። 

# ሁኔታዊ ዕድል

ዕድል (ፕሮባቢሊቲ) በ ዜሮ እና በ አንድ (ሁለቱንም ጨምሮ) መካከል ባለ ቁጥር የሚገለጽ የ አንድ ግምት የመፈጸም አቅሙ ነው። አንድ ግምቱ መቶ በመቶ የሚፈጸም መሆኑን ሲያሳይ ዜሮ ደግሞ ግምቱ ምንም የመሆን እድል የለውም ማለት ነው። በሁለቱ መካከል ያሉት ቁጥሮች ግምቱ የመሆን እድሉን ይገልጻሉ። ምሳሌ 0.35 ወይም 0.50 ግምት የመፈጸም አቅሙ 35 በመቶ እና 50 በመቶ መሆኑን ይገልጻሉ። 

ሁኔታዊ ዕድል የምንለው በአንጻሩ የቀደመ መረጃን ተንተርሶ የሚሰራ ነው። ለምሳሌ በ ፲ የፓለቲካ ፓርቲዎች ለምርጫ ቢወዳደሩ እያንዳንዱ ፓርቲ ፩/፲ ወይም 0.1 የማሸነፍ ዕድል አላቸው። ነገር ግን የመሬት፣ የስራ ፈጠራ እና የግብርና ፓሊሲያቸው በህዝብ ተቀባይነት ያላቸው ፓርቲዎች የማሸነፍ ዕድላቸው በ 0.3 እጅ የተሻለ ነው የሚል የቀደመ ጥናት ቢኖር፣ እነዚህን ሁኔታዎች በፖሊሲው በአግባቡ መመለስ የቻለ በምርጫ የማሸነፍ እድሉ ከ 0.1 ወደ 0.4 ከፍ ይላል ማለት ነው።   

*** 
ሁኔታዊ እድል በ p(A/B) ይገለጻል። ይህም ማለት የ "ኤ" ዕድል ቢ እውነት ሲሆን ማለት ነው።  ከላይ በተገለጸው ምሳሌ መሰረት ፓርቲ ሀ (የመሬት፣ የስራ ፈጠራ እና የግብርና ፓሊሲ በህዝብ ተወዳጅ የሆነለት) በምርጫ የማሸነፍ እድሉ ማለት ነው። ኤ (ፓርቲው በምርጫ የማሸነፍ ዕድሉ) ቢ ደግሞ የተዘረዘሩት የመሬት፣ የስራ ፈጠራ እና የግብርና ፓሊሲ ሁኔታዎች ናቸው።  
*** 

# ቅንጅታዊ ዕድል

ይህ ማለት ሁለት ነገሮች ተያይዘው የመሆን እድላቸው ማለት ነው። ሲገለጽም p(A and B) ተብሎ ይገለጻል። ምሳሌ ሁለት ሳንቲሞች ወደ ላይ ብንወረውር እና የመጀመሪያው ሳንቲም ጎፈር፣ ሁለተኛው ሳንቲም ጎፈር የመሆን እድላቸው ማለት ነው። 

*** 
ዕ(ጎ)(የመጀመሪያው ሳንቲም) = (ጎፈር የመሆን እድሉ) = ፩/፪ = 0.5 

ዕ(ጎ)(ሁለተኛው ሳንቲም) = (ጎፈር የመሆን እድሉ) = ፩/፪ = 0.5

p(A and B) = p(A) $\times$ P(B) = 0.5 $\times$ 0.5 = 0.25 (ድርጊቶቹ እራሳቸውን የቻሉ ስለሆኑ)

ድርጊቶች ግን እራሳቸውን ችለው የሚቆሙ ላይሆኑ ይችላሉ። ምሳሌ ሁለት ተከታታይ ቀናት የምዝነብ ዕድሉ፤ ዛሬ እንደዘነብ ካወቅን፤ ነገ የመዝነብ ዕድሉን መገመት እንችላለን። ባጠቃላይ። 

p(A and B) = p(A) $\times$ P(B/A) ዛሬ ከዘነበ፤ ነገም የመዝነብ ዕድሉ ከ 0.25 ከፍ ያለ ዕድል ይኖረዋል።  
*** 

# ቤዚያን ቴረም

እንበልና ሁለት ከረሜላ የያዙ ሳህኖች አሉ። ሳህን-፩ እና ሳህን-፪። ሳህን-፩ 30 ደስታ ከረሜላ እና 10 ናና ከረሜላ ይዟል። ሳህን-፪ 20 ደስታ ከረሜላ እና 20 ናና ከረሜላ ይዟል እንበል። ጥያቄው በዘፈቀደ እጃችንን ወደሳህኖቹ ብንሰድ እና ደስታ ከረሜላ ብናነሳ፣ ደስታ ከረሜላው ከ ሳህን-፩ የመሆኑ እድሉ ምን ያህል ነው ብለን ብንጠይቅ መልሱን ለመገመት እንቸገራለን። 

ምክኒያቱም ጥያቄው ሁኔታዊ ዕድል ሲሆን እሱም P(ሳህን-፩/ደስታ ከረሜላ) ነው። P(ደስታ ከረሜላ/ሳህን-፩) የመሆን እድሉ አይደለም። ደስታ ከረሜላን ከሳህን-፩ የማንሳት ዕድል - `P(ደስታ ከረሜላ/ሳህን-፩)` እና ከሳህን-፩ ላይ ደስታ ከረሜላን የማንሳት እድል - `P(ሳህን-፩/ደስታ ከረሜላ)` ይለያያሉ። 

ልዩነቱም እንደሚከተለው ነው። 
* P(ደስታ ከረሜላ/ሳህን-፩) - ሳህን-፩ ቀርቦልን ደስታ ከረሜላ የማንሳት እድል = 30/40 = 3/4 
* P(ሳህን-፩/ደስታ ከረሜላ) - ሳህኖቹ ቀረቡልን (ሁለቱም) ደስታ ከረሜላ አነሳን፤ ከሳህን-፩ ላይ የመሆን እድሉ ነው ጥያቄው (ይሔንን ነው ቤዚያን ቴረም የሚመልስልን) በመደበኛ ሁኔታዊ ዕድል መመለስ የማንችለውን። 

*** 
P(A and B) = P(B and A) (በማንኛውም ክስተት ተለዋዋጭ የሒሳብ ባህሪ ነው)

P(A and B) = P(A) $\times$ P(B/A) (ከላይ ተመልከቱ) 

P(B and A) = P(B) $\times$ P(A/B) 

P(B)$\times$P(A/B) = P(A)$\times$P(B/A) (የሁለቱ ውጤት ነው) 

P(A/B) = $\frac{P(A) \times P(B/A)}{P(B)}$


***

ከላይ ያለውን ቀመር በመጠቀም P(ሳህን-፩/ደስታ ከረሜላ) ዕድል መገመት እንችላለን ማለት ነው። የሚያስፈልጉን 

* P(ሳህን-፩) = ሁለት ሳህኖች ስለሆኑ ያሉን ዕድሉ = 1/2 ነው። 
* P(ደስታ ከረሜላ/ሳህን-፩) = ሳህን-፩ ላይ 40 ከረሜላዎች አሉ፤ 30 ደስታ ከረሜላ ናቸው። 30/40 = 3/4 
* P(ደስታ ከረሜላ) = ሁለቱም ሳህኖች ውስጥ 50 ደስታ ከረሜላዎች አሉ። አጠቃላይ 80 ከረሜላዎች ናቸው ያሉት። በመሆኑም P(ደስታ ከረሜላ) = 50/80 = 5/8 ይሆናል ማለት ነው። 
* P(ሳህን-፩/ደስታ ከረሜላ) = (1/2 * 3/4)/5/8 = (3/8)/(5/8) = 3/5 ይሆናል ማለት ነው። 


ዊሂብ (ዳታ) ላይ ተመስርተን የምናደርገውን መላምት (ሀይፓተሲስ) እንደሚከተለው መግለጽ እንችላለን። 
*** 
$P(H/D) = \frac{P(H)\times P(D/H)}{P(D)}$
*** 

* $P(H/D)$ የምንቀምረው ስለሆነ *ፓስተሪየር* ወይም *ተከታይ* እንደማለት ነው። 
* $P(H)$ መላምቱ ምንም አይነት ውሂብ (ዳታ) ከማየታችን በፊት ማለት ነው። *ፕራየር* ወይም *ቀዳሚ* እንደማለት ነው። 
* $P(D/H)$ የውሂቡ (ዳታው) ዕድል በመላምቱ *ላይክሊሁድ* ወይም *የመሆን ዕድል* ይባላል 
* $P(D)$ በማንኛውም መላምት የውሂቡ ዕድል *ኖርማላይዚንግ ኮንስታንት* ወይም *በመደበኛነት ቋሚ* እንደማለት ነው። 

# ምሳሌ

ቤዚያንን የሚጠቀም አንድ ምሳሌ እንመልከት እና ይህንን ክፍል እናጠቃል። ምሳሌው እንደሚከተለው ነው። 

***
ከረሜላ አምራች ድርጅቱ (ድርጅት ሀ እንበለው) የከረሜላ ቀለሞቹን እንደሚከተለው አዋቅሯል። 30% ቡናማ፣ 20% ቢጫ፣ 20% ቀይ፣ 10% አረንጓዴ፣ 10% ብርቱካናማ እና የቀረውን 10% ጠይም ቀለም አድርጎ ሲጠቀም ነበር። ከ 1995 ወዲህ ግን 24% ሰማያዊ፣ 20% አረንጓዴ፣ 16% ብርቱካናማ 14% ቢጫ፣ 13% ቀይ እና 13% ቡናማ በማድረግ ከረሜላዎቹን እየሰራ ይገኛል። ጥያቄው አንድ ሰው እነዚህን  ከረሜላዎች በሁለት ቦርሳ (ከ1995 በፊት የተመረቱትን በአንድ ቦርሳ፣ ከ 1995 ቦኋላ የተመረቱትን በሌላ ቦርሳ) ይዞ ከነበረበት በዘፈቀደ አንድ ቢጫ አንድ አረንጓዴ ከረሜላ ከሁለቱም ቦርሳዎች አንድ አንድ አንስቶ ቢሰጠን፣ ቢጫው ከ 1995 ከተመረቱት በፊት የመምጣት ዕድሉ ምን ያህል ነው? ጥያቄውን በ ሰንጠረዥ ለማገዝ፡

```python
import pandas as pd
ሀ = {'ማውጫ': ['ቡናማ', 'ቢጫ', 'ቀይ', 'አረንጓዴ', 'ብርቱካናማ', 'ጠይም', 'ሰማያዊ'],
     'ከ1995 በፊት(በመቶኛ)': [30, 20, 20, 10, 10, 10, 0],
     'ከ1995 በኋላ (በመቶኛ)': [13, 14, 13, 20, 16, 0, 24]}

#ውሂብ = [[1, 2], [3, 4]]
#ማውጫ = 
#pd.DataFrame(, columns=["ከ1995 በፊት", "ከ1995 በኋላ"])
pd.DataFrame(ሀ).set_index('ማውጫ').reset_index()
```

ጥያቄው ቢጫው ከረመላ ከ 1995 በፊት ከሆነው ቦርሳ የመምጣት ዕድሉ ምን ያህል ነው? ሁለት መላምቶች ይኖሩናል። 
* ቢጫው ከ 1995 በፊት ካለው ቦርሳ እና አረንጓዴው ከ 1995 በኋላ ካለው ቦርሳ (ሀ) 
* ቢጫው ከ 1995 ቦኋላ ካለው ቦርሳ እና አረንጓዴው ከ 1995 በፊት ካለው ቦርሳ (ለ)

ፕራየር (የ መላምት ሀ እና የመላምት ለ) ዕድል እኩል ነው። 1/2 

በተሰጠው ውሂብ ላይ ተመስርቶ መላምት ሀ የመሆን ዕድሉ (አንድ ቢጫ እና አንድ አረንጓዴ ከ ቦርሳ ሀ የመሆን ዕድሉ) = 20/100 * 20/100 =4/100=1/25፤ የመላምት ለ የመሆን ዕድሉ 14/100 * 10/100 = 7/500

የመደበኛ ቋሚው (ኖርማላይዚንግ ኮንስታንት) የምንለው = (1/2 * 1/25) + (1/2 * 7/500) = (1/50) + (7/1000)= 27/1000

P(ሀ/በተሰጠን ዊሂብ ተመስርቶ) = (1/2 * 1/25) /(27/1000)= 20/27 

P(ለ/በተሰጠን ዊሂብ ተመስርቶ) = (1/2 * 7/500)/(27/1000) = 7/27 

# ማጠቃለያ 

ሁኔታዊ ዕድሎችን በሚመለከት የሚነሱ ጥያቄዎችን ለመመለስ ቤዚያን ቴረም አማራጮችን ይሰጣል። ከቀመሩ እንደምንመለከተው $P(H/D) = \frac{P(H)\times P(D/H)}{P(D)}$ ከተሰጠን መረጃ ተነስተን ልናገኝ ወይም ልንቀምራቸው የምንችላቸውን ወስደን ተፈላጊውን የመቀመር አማራጭ ይሰጣል። በሚቀጥለው ክፍል ፓይተንን በመጠቀም ቤዚያንን እንዴት እንደምንቀምር እንመለከታለን።
